## The code snippet below is used to extract  data from  "Amazon website", i extracted the 'Rank No of 100 best selling books for the given years(2009-2019) ', 'Title of books' and 'Books type'.

# Import relevant libraries

```python
import requests # this is used to get the data from the website
from bs4 import BeautifulSoup as bs
import time
import csv # This will be used later to save the file as "csv"
import os  #os means operating system
```

## We scrape in large quantity to make use of the data and save as "csv" for storage.

```python
pages = [1, 2] #we specify the number of pages we want to loop (2 pages)

#We save as csv, specify the folder we want it to be stored. 'a' means append as we merge 3 columns,'newline' to give it space
with open('C:/Users/HP/Jupyter_folder/Hamoye/B_2009.csv', 'a',encoding='utf-8', newline='') as f_output:
    csv_print = csv.writer(f_output)
    
    file_is_empty = os.stat('C:/Users/HP/Jupyter_folder/Hamoye/B_2009.csv').st_size == 0
    if file_is_empty:          #if file is empty, just print the column head alone 
        csv_print.writerow(['Rank_No', 'Book_title', 'Book_type'])
        
#write the code in a for loop to look into other book listng
    for page in pages:
        source = requests.get('https://www.amazon.com/gp/bestsellers/2009/books/ref=zg_bsar_pg_2?ie=UTF8&pg=2'.format(page)).text
        soup = bs(source, 'lxml')# convert into a beautifulsoup object and parse as lxml which is preferable.

#try except is used to skip missing part of the script so as not to stop it.
        for books in soup.find_all('li', class_='zg-item-immersion'):
         
            try:
                rank = books.find(class_="zg-badge-text").get_text().strip()
            except Exception as e:
                rank = None
            print('Rank_No:',rank)
    
            try:
                title = books.span.a.get_text().strip()
            except Exception as e:
                title = None
            print('Book_title:',title)
    
            try:
               types = books.find('span', class_='a-size-small a-color-secondary').get_text().strip()
            except Exception as e:
                types = None
            print('Book_type:',types)
          
    
            print('---------------') #this is used for easy damarcation.
    
            time.sleep(2) #it will take 2 seconds before the next page load, this is to
                  #regulate the scraping and not go too hard on the website.
    
            csv_print.writerow([rank, title, types])
```
